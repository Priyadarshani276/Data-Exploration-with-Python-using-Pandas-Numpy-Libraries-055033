**Outlier Detection**

Outliers are data points that differ significantly from other observations. Identifying outliers is critical, as they can skew results and lead to inaccurate conclusions. Common methods for detecting outliers include:

* Z-score: A statistical measure that describes a value's relation to the mean of a group of values. A high Z-score (typically above 3 or below -3) indicates an outlier.

* Interquartile Range (IQR): The IQR is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data. Outliers are often defined as values that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.

Outlier Detection

Outliers are data points that differ significantly from other observations. Identifying outliers is critical, as they can skew results and lead to inaccurate conclusions. Common methods for detecting outliers include:

●	Z-score: A statistical measure that describes a value's relation to the mean of a group of values. A high Z-score (typically above 3 or below -3) indicates an outlier.

●	Interquartile Range (IQR): The IQR is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data. Outliers are often defined as values that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.

```python
#96. Perform an ANOVA Test to Compare the Means Across Multiple Groups
from scipy.stats import f_oneway

# Ensure 'Category' and 'Value' columns exist
if 'Category' in sample_data.columns and 'Value' in sample_data.columns:
    groups = [sample_data[sample_data['Category'] == group]['Value'].dropna() for group in sample_data['Category'].unique()]
    
    # Ensure there are more than one group and enough observations
    if all(len(group) > 1 for group in groups) and len(groups) > 1:
        # Perform ANOVA
        anova_result = f_oneway(*groups)
        print(f"ANOVA: F-statistic = {anova_result.statistic}, p-value = {anova_result.pvalue}")
    else:
        print("Not enough groups or observations to perform ANOVA.")
else:
    print("'Category' or 'Value' column does not exist in the DataFrame.")
```

```python
#97.  Test the Variance Using Bartlett's Test
from scipy.stats import bartlett

# Ensure 'Category' and 'Value' columns exist
if 'Category' in sample_data.columns and 'Value' in sample_data.columns:
    groups = [sample_data[sample_data['Category'] == group]['Value'].dropna() for group in sample_data['Category'].unique()]
    
    # Ensure groups have enough observations
    if all(len(group) > 1 for group in groups):
        # Perform Bartlett's test
        bartlett_result = bartlett(*groups)
        print(f"Bartlett's Test: statistic = {bartlett_result.statistic}, p-value = {bartlett_result.pvalue}")
    else:
        print("Not enough observations in groups for Bartlett's test.")
else:
    print("'Category' or 'Value' column does not exist in the DataFrame.")
```

```python
#98. Test for Normality Using the Shapiro-Wilk Test
from scipy.stats import shapiro

# Ensure 'Value' column exists
if 'Value' in sample_data.columns:
    value_column = sample_data['Value'].dropna()
    
    if len(value_column) > 2:  # Shapiro-Wilk test requires more than 2 samples
        # Perform the Shapiro-Wilk test
        shapiro_test = shapiro(value_column)
        print(f"Shapiro-Wilk Test: statistic = {shapiro_test.statistic}, p-value = {shapiro_test.pvalue}")
    else:
        print("Not enough data to perform Shapiro-Wilk test.")
else:
    print("'Value' column does not exist in the DataFrame.")
```

```python
#99. Test for Normality Using the Kolmogorov-Smirnov Test
from scipy.stats import kstest
import numpy as np

# Ensure 'Value' column exists
if 'Value' in sample_data.columns:
    value_column = sample_data['Value'].dropna()
    
    if len(value_column) > 2:
        # Normalize the data for KS test (optional)
        norm_value_column = (value_column - np.mean(value_column)) / np.std(value_column)
        
        # Perform the Kolmogorov-Smirnov test
        kstest_result = kstest(norm_value_column, 'norm')
        print(f"Kolmogorov-Smirnov Test: statistic = {kstest_result.statistic}, p-value = {kstest_result.pvalue}")
    else:
        print("Not enough data to perform Kolmogorov-Smirnov test.")
else:
    print("'Value' column does not exist in the DataFrame.")
```

```python
# 105. Polynomial Regression with Interaction Terms


import pandas as pd
from sklearn.preprocessing import PolynomialFeatures, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Encode 'Shipping_Method' as numeric
if 'Shipping_Method' in sample_data.columns:
    label_encoder = LabelEncoder()
    sample_data['Shipping_Method'] = label_encoder.fit_transform(sample_data['Shipping_Method'])

# Ensure 'Quantity', 'Shipping_Method', and 'Value' columns exist
if 'Quantity' in sample_data.columns and 'Shipping_Method' in sample_data.columns and 'Value' in sample_data.columns:
    df_clean = sample_data[['Quantity', 'Shipping_Method', 'Value']].dropna()
    X = df_clean[['Quantity', 'Shipping_Method']]
    y = df_clean['Value']
    
    if len(X) > 1 and len(y) > 1:
        # Polynomial transformation with interaction terms
        poly = PolynomialFeatures(degree=2, interaction_only=True)
        X_poly = poly.fit_transform(X)

        # Initialize and fit the model
        model = LinearRegression()
        model.fit(X_poly, y)

        # Predict and evaluate
        y_pred = model.predict(X_poly)
        mse = mean_squared_error(y, y_pred)
        r2 = r2_score(y, y_pred)

        print(f"Polynomial Regression with Interaction Terms (Degree 2):")
        print(f"Coefficients: {model.coef_}")
        print(f"Intercept: {model.intercept_}")
        print(f"Mean Squared Error: {mse}")
        print(f"R^2 Score: {r2}")
    else:
        print("Not enough data in 'Quantity', 'Shipping_Method', or 'Value' to perform the regression.")
else:
    print("'Quantity', 'Shipping_Method', or 'Value' column does not exist in the DataFrame.")
```

```python
#109. Find the correlation between two categorical variables
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    crosstab = pd.crosstab(x, y)
    chi2_stat = chi2_contingency(crosstab)[0]
    n = crosstab.sum().sum()
    k = min(crosstab.shape) - 1
    return np.sqrt(chi2_stat / (n * k))

# Ensure the columns exist
if 'Category' in sample_data.columns and 'Product' in sample_data.columns:
    correlation = cramers_v(sample_data['Category'], sample_data['Product'])
    
    print(f"Cramér's V correlation between 'Category' and 'Product':")
    print(correlation)
else:
    print("'Category' or 'Product' does not exist in the DataFrame.")
```
